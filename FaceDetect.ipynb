```python
import cv2
import dlib
import matplotlib.pyplot as plt
```

### face detection from a picture

```python
class FaceDetector:
    def __init__(self, image_path):
        self.image_path = image_path
        self.image = cv2.imread(image_path)

        if self.image is None:
            raise ValueError(f"Unable to read the image from {image_path}")

    def detect_faces(self, scaleFactor, minNeighbors, minSize):
        # Instantiate the face detector object
        fd = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

        image_gray = cv2.cvtColor(self.image, cv2.COLOR_BGR2GRAY)
        faces = fd.detectMultiScale(
            image_gray,
            scaleFactor=scaleFactor,
            minNeighbors=minNeighbors,
            minSize=minSize
        )

        for x, y, w, h in faces:
            cv2.rectangle(self.image, (x, y), (x + w, y + h), (127, 255, 0), 3)

    def display_image(self):
        img_rgb = cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(20, 10))
        plt.imshow(img_rgb)
        plt.axis('off')
        plt.show()
```

```python
image_path = '12.jpeg'
face_detector = FaceDetector(image_path)

face_detector.detect_faces(scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))
face_detector.display_image()
```
